```python
import numpy as np
```

```python
Tbase_gu_stage = [
    13.37766,
    13.37766,
    13.37766,
    9.784431
]
Tthresh_gu_stage = [
    38.50,
    47.61,
    47.39,
    316.376
]
gu_pheno_tts = np.zeros(1)
gu_stage = np.zeros(1)
TM = np.random.randint(9, 21, 200)
```

```python
for t in TM:
    for stage, thresh, base in zip(reversed(range(len(Tthresh_gu_stage))), reversed(Tthresh_gu_stage), reversed(Tbase_gu_stage)):
        in_stage = (gu_stage >= stage) & (gu_stage < stage + 1)
        if not np.any(in_stage):
            continue 
        gu_pheno_tts[in_stage] += max(0, t - base)
        share = gu_pheno_tts[in_stage] / thresh
        gu_stage[in_stage] = np.where(share > 1., stage + 1., stage + share)
        gu_pheno_tts[share > 1.] = 0.
        print(stage, in_stage, gu_stage, gu_pheno_tts, t, t - base, max(0, t - base))
```
